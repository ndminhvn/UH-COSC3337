## Plotting Numpy Arrays

Plot a test graph

```python
x = np.linspace(0, 25) # x-axis range
y = np.cos(x) # y-axis range
plt.plot(x, y) # plot the graph
plt.title('y = cos(x) graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show() # display the graph
```

Plot legends

```python
plt.plot(x, np.sin(x), label='y=sin(x)')
plt.plot(x, np.cos(x), label='y=cos(x)')
plt.title('Sine and Cosine Functions ')
plt.xlabel("x-axis")
plt.ylabel("y-axis")
# Assigns a color to each plot along with it's corresponding label name to display in a legend.
plt.legend()
plt.show()
```

## Subplots

1d array

```python
# one figure object with 1x2=2 axes
fig, axes = plt.subplots(nrows=1, ncols=3)

# plotting sin(x) in axes[0]. left image, or 1st element/axis in the axes array.
axes[0].plot(x, np.sin(x))
# setting the title of axes[0] to be y=sin(x)
axes[0].set_title('y = sin(x)')

# plotting cos(x) in axes[1]. right image, or 2nd element/axis in the axes array.
axes[1].plot(x, np.cos(x))
# setting the title of axes[1] to be y=cos(x)
axes[1].set_title('y = cos(x)')

# example
axes[2].plot(x, np.sin(x) + np.cos(x))
axes[2].set_title('y = sin(x) + cos(x)')

plt.show()
```

2d array

```python
x = np.linspace(1, 21)

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(14, 8))

# Customize the graphs
axes[0][0].plot(x, x)
axes[0][0].set_title('O(n)')

axes[0][1].plot(x, x*2)
axes[0][1].set_title('O(n^2)')

axes[1][0].plot(x, np.log(x))
axes[1][0].set_title('O(log(n))')

axes[1][1].plot(x, np.log(x)*x)
axes[1][1].set_title('O(nlog(n))')

plt.show()
```

## Custom Axes Range

Examples

```python
x = np.linspace(1, 5, 20)

fig, axes = plt.subplots(nrows=1, ncols=3)

# we can also plot multiple graphs using plt.plot(x1, y1, x2, y2).
axes[0].plot(x, x**2, x, np.log(x))
axes[0].set_title('default axis range')

# tight happens to be the same here, but this will come in handy when you want to automatically align axes in a
# fitted manner.
axes[1].plot(x, x**2, x, np.log(x))
axes[1].axis('tight')
axes[1].set_title('tight axis range')

# notice the zoom in that happened.
axes[2].plot(x, x**2, x, np.log(x))
axes[2].set_ylim([0, 40])
axes[2].set_xlim([2, 3])
axes[2].set_title("custom axes range");

plt.show()
```

Additional arguments for plt.plot()

```python
# Custom list of graphs

x = np.linspace(0, 5, 11)

plt.plot(x, 1-x, color="red", linewidth=0.25)
plt.plot(x, 2-x, color="red", linewidth=0.50)
plt.plot(x, 3-x, color="red", linewidth=1.00)
plt.plot(x, 4-x, color="red", linewidth=2.00)

# possible linestype options -, , -., :, steps
plt.plot(x, 5-x, color="green", lw=3, linestyle='-')
plt.plot(x, 6-x, color="green", lw=3, ls='-.')
plt.plot(x, 7-x, color="green", lw=3, ls=':')

# custom dash (not something we'll do often)
line, = plt.plot(x, 0*x+8, color="black", lw=1.50)
line.set_dashes([5, 10, 15, 10]) # format: line length, space length, ...

# possible marker symbols: marker = '+', 'o', '*', 's', ',', '.', '1', '2', '3', '4', ...
plt.plot(x, x+9, color="blue", lw=3, ls='-', marker='+')
plt.plot(x, x+10, color="blue", lw=3, ls='--', marker='o')
plt.plot(x, x+11, color="blue", lw=3, ls='-', marker='s')
plt.plot(x, x+12, color="blue", lw=3, ls='--', marker='1')

# marker size and color
plt.plot(x, x+13, color="purple", lw=1, ls='-', marker='o', markersize=2)
plt.plot(x, x+14, color="purple", lw=1, ls='-', marker='o', markersize=4)
plt.plot(x, x+15, color="purple", lw=1, ls='-', marker='o', markersize=8, markerfacecolor="red")
plt.plot(x, x+16, color="purple", lw=1, ls='-', marker='s', markersize=8, 
        markerfacecolor="yellow", markeredgewidth=3, markeredgecolor="green");
```

Turn each unique pokemon type into a bar plot

```python
pokemon_df['Type 2'].value_counts()
```

Make bar plot

```python
# collect index names (left side of .value_counts())
type_2_names = pokemon_df['Type 2'].value_counts().index
# collect corresponding value counts (right side of .value_counts())
type_2_values = pokemon_df['Type 2'].value_counts().values
# plt.bar(x, y)
plt.bar(type_2_names, type_2_values)
plt.title('Type 2 Pokemon value counts')
plt.show()
```

```python
# collect index names (left side of .value_counts())
type_1_names = pokemon_df['Type 1'].value_counts().index
# collect corresponding value counts (right side of .value_counts())
type_1_values = pokemon_df['Type 1'].value_counts().values
# plt.bar(x, y)
plt.bar(type_1_names, type_1_values)
plt.title('Type 1 Pokemon value counts')
plt.show()
```

Pie chart

```python
# colors we'll use for each patch/wedge
colors = ['lightgreen', '#FFDD44', 'lightblue', 'lightcoral','red']
patches, texts = plt.pie(type_1_values[:5], colors=colors, shadow=True, startangle=90)
plt.legend(patches, type_1_names[:5])
# will center our pie chart, but not necessary
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Collect the first few pokemon names in our DataFrame along with their attributes - stack plot

```python
x = pokemon_df['Name'][:10]
y1 = pokemon_df['HP'][:10]
y2 = pokemon_df['Attack'][:10]
y3 = pokemon_df['Defense'][:10]
y4 = pokemon_df['Speed'][:10] # custom
```

```python
labels = ["HP ", "Attack", "Defense", "Speed"]

plt.stackplot(x, y1, y2, y3, y4)
plt.legend(loc='upper left', labels=labels)
# rotates the xlabels to appear vertical rather than horizontal by default. 
plt.xticks(rotation=90)
plt.show()
```

subplots with above data

```python
# custom
# Create 4 subplots sharing y axis
fig, axes = plt.subplots(nrows=4, ncols=1, sharey=True)

# 'ko-, r.-, and : are just line styles and optional.'
# notice how you can also set multiple things together in one single command rather than
# creating separate .set_title, .set_ylabel, .set_xlabel, etc. commands like we were doing earlier.
axes[0].plot(x, y1, 'ko-')
axes[0].set(title='4 subplots', ylabel='HP')

axes[1].plot(x, y2, 'r.-')
axes[1].set(ylabel='Attack')

axes[2].plot(x, y3, '-')
axes[2].set( ylabel='Defense')

axes[3].plot(x, y4, '-.')
axes[3].set(xlabel='Pokemon', ylabel='Speed')

plt.show()
```
